```python
df_old = pd.read_csv('benchmark_old_column_names.csv', delimiter=',')

# Drop all columns that start with 'own_'
df_old = df_old.loc[:, ~df_old.columns.str.startswith('own_')]

# Replacing names of all iteration-based columns
replace_dict = {
    'l2_delta1': 'l2_norm_delta_image1_perturbed',
    'l2_delta2': 'l2_norm_delta_image2_perturbed',
    'l2_delta12': 'l2_norm_delta_avg_images_perturbed',
    'l0_delta1': 'l0_norm_delta_image1_perturbed',
    'l0_delta2': 'l0_norm_delta_image2_perturbed',
    'l0_delta12': 'l0_norm_delta_avg_images_perturbed',
    'l_inf_delta1': 'l_inf_norm_delta_image1_perturbed',
    'l_inf_delta2': 'l_inf_norm_delta_image2_perturbed',
    'l0_inf_delta12': 'l_inf_norm_delta_avg_images_perturbed',
    'epe_gt': 'epe_to_ground_truth',
    'epe_target': 'epe_to_target',
    'epe_orig_preds': 'epe_clean_predictions_to_ground_truth',
    'cosim_target': 'cosine_similarity_to_target',
    'cosim_orig_preds': 'cosine_similarity_to_clean_predictions',
    'cosim_ground_truth': 'cosine_similarity_to_ground_truth',
}


df_old.columns = [col for col in df_old.columns]
for old_part, new_part in replace_dict.items():
    df_old.columns = [col.replace(old_part, new_part) for col in df_old.columns]

rename_dict = {
    'boxconstraint': 'pcfa_boxconstraint',
    'attack': 'threat_model',
    'norm': 'lp_norm',
    'loss': 'loss_function',
    'epe': 'epe_final_prediction',
    'alph': 'weather_alph_motionoffset',
    'do': 'weather_do_motionblur', 
    'num': 'weather_num_flakes',
    'time': 'model_proposal_date',
    'target': 'attack_target',
    'optim': 'optimization_target',
    'cosim': 'cosine_similarity_final_prediction',
    'epe_ground_truth': 'epe_final_prediction_to_ground_truth',
    'epe_initial_to_zero': 'epe_clean_prediction_to_zero_target',
    'epe_initial_to_negative': 'epe_clean_prediction_to_negative_target',
    'severity': 'common_corruption_severity',
    'name': 'common_corruption_name',
    '3dcc_intensity': '3d_common_corruption_severity',
    '3dcc_corruption': '3d_common_corruption_name',
    'optimizer': 'weather_optimizer', 
    'steps': 'weather_steps', 
    'learn': 'weather_learn_transparency', 
    'rendering': 'weather_rendering_method', 
    'transparency': 'weather_transparency_scale', 
    'depth': 'weather_depth_check_differentiable', 
    'scene': 'weather_scene_scale', 
    'recolor': 'weather_recolor', 
    'motionblur': 'weather_motionblur_samples', 
    'flakesize': 'weather_flakesize_max', 
    'constant': 'weather_constant_transparency', 
    'motion': 'weather_motion_y', 
    'flake': 'weather_flake_b', 
    'frame': 'weather_frame_per_scene', 
    'no': 'weather_no_flake_dat', 
    'lr': 'weather_lr', 
    'unregistered': 'weather_unregistered_artifacts', 
}

df_old.rename(columns=rename_dict, inplace=True)

df_old.to_csv('benchmark.csv', index=False)
```
